4. Create a random numpy array of size 6x3 and transpose it

```python
# Question 4
array3 = np.random.random((6,3))
print("Original random 6x3 Array:")
print(array3)
```

```python
transposedArray3 = array3.T
print("Transposed Array:")
print(transposedArray3)
```

5. Create two random numpy array of size n x p and p x m and then compute the dot product

```python
# Question 5
n = 2
p = 3
m = 4
arrayA = np.random.random((n,p))
print("Array A (n x p):")
print(arrayA)
```

```python
arrayB = np.random.random((p,m))
print("Array B (p x m):")
print(arrayB)
```

```python
dotProduct = np.dot(arrayA,arrayB)
print("Dot Product:")
print(dotProduct)
```
